```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
purchase_behavior = pd.read_csv('purchase_behavior.csv')
transaction_data = pd.read_csv('transaction_data.csv')
```

```python
transaction_data["DATE"] = pd.to_datetime(transaction_data["DATE"], origin="1899-12-30", unit="D")
```

```python
merged_data = pd.merge(transaction_data, purchase_behavior, on='LYLTY_CARD_NBR', how='inner')
```

### 1. Top 3 Profitable Products

To find the most profitable product, Calculate the total revenue generated by each product and then sort them in descending order. The top products in this list represent the most profitable ones.

**Formula:**  
Total Revenue = ∑ Sales of Each Product

```python
product_revenue = transaction_data.groupby("PROD_NAME")["TOT_SALES"].sum().reset_index()

top_3_products = product_revenue.sort_values(by="TOT_SALES", ascending=False).head(3)

print("Top 3 Most Profitable Products:")
top_3_products
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TOT_SALES', y='PROD_NAME', data=top_3_products, palette='viridis')
plt.title('Top 3 Most Profitable Products')
plt.xlabel('Total Sales ($)')
plt.ylabel('Product Name')
plt.show()
```

### 2. Most Loyal Customers

To identify the most loyal customers, group transactions by 'LIFESTAGE' and 'PREMIUM_CUSTOMER'. For each segment, calculate:

Total transactions (sum of all purchases)

Unique customers (count of distinct loyalty card numbers).
Divide total transactions by unique customers to compute transactions per customer, which measures loyalty independent of segment size. The segment with the highest ratio is the most loyal, as it reflects repeat purchases per individual. This method ensures smaller but highly engaged groups aren’t overshadowed by larger, less loyal segments.

```python
loyalty_analysis = (merged_data.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER']).agg(total_transactions=('TXN_ID', 'count'), unique_customers=('LYLTY_CARD_NBR', 'nunique')).reset_index())

loyalty_analysis['transactions_per_customer'] = (loyalty_analysis['total_transactions'] / loyalty_analysis['unique_customers'])

top_segment = loyalty_analysis.sort_values('transactions_per_customer', ascending=False).head(1)

print("\nMost Loyal Customer Segment:")
top_segment
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='transactions_per_customer', y='LIFESTAGE', hue='PREMIUM_CUSTOMER', 
    data=loyalty_analysis.sort_values('transactions_per_customer', ascending=False).head(5),palette='coolwarm')

plt.title('Top Customer Segments by Transaction Frequency')
plt.xlabel('Average Transactions per Customer')
plt.ylabel('Life Stage')
plt.legend(title='Customer Type')
plt.show()
```
